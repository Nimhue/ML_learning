```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```python
alice_frequency = nltk.FreqDist(alice)
```

```python
alice_frequency
```

```python
alice_frequency["Rabbit"]
```

```python
alice_frequency["Alice"]
```

```python
alice_frequency.most_common()
```

```python
alice_frequency.most_common(10)
```

```python
alice_frequency.hapaxes()
```
